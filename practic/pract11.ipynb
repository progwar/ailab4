```python
import pandas as pd
import numpy as np

date_range = pd.date_range(start='2024-03-01', periods=10, freq='D')

random_values = np.random.randint(10, 101, 10)
s = pd.Series(random_values, index=date_range)

print("Временной ряд (1-10 марта 2024):")
print(s)

print("Данные за 5-8 марта:")
print(s['2024-03-05':'2024-03-08'])
```
