```python
import pandas as pd
import numpy as np

data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
s = pd.Series(data)

print("Исходный Series:")
print(s)

has_nan = s.isna().any()
print(f"Есть ли пропущенные значения? {has_nan}")

nans = s[s.isna()].index
print(f"Индексы пропущенных значений: {nans}")
```
