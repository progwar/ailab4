```python
import pandas as pd

s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

print("s1:")
print(s1)

print("s2:")
print(s2)

s_sum = s1 + s2
print("Результат сложения s1 + s2:")
print(s_sum)

print("Объяснение появления NaN:")
print("Общие индексы: ['b', 'c', 'd']")
print("Индексы только в s1: ['a'] -> NaN")
print("Индексы только в s2: ['e'] -> NaN")

s_sum_filled = s_sum.fillna(0)
print("Результат после замены NaN на 0:")
print(s_sum_filled)
```
