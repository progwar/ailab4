```python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50, 60], index=['A', 'B', 'A', 'C', 'D', 'B'])

print("Исходный Series с повторяющимися индексами:")
print(s)

is_unique = s.index.is_unique
print(f"Индексы уникальны? {is_unique}")

if not is_unique:
    print("Индексы не уникальны. Выполняем группировку и суммирование:")
    s_grouped = s.groupby(s.index).sum()
    print(s_grouped)
else:
    print("Индексы уникальны. Группировка не требуется.")
```
