```python
import pandas as pd
import numpy as np

random_numbers = np.random.randint(50, 151, 20)
s = pd.Series(random_numbers)

print("Созданный Series:")
print(s)

print("Основные статистические характеристики:")
print(f"Сумма: {s.sum()}")
print(f"Среднее: {s.mean():.2f}")
print(f"Минимальное значение: {s.min()}")
print(f"Максимальное значение: {s.max()}")
print(f"Стандартное отклонение: {s.std():.2f}")
```
