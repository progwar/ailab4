```python
import pandas as pd
import numpy as np

data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
s = pd.Series(data)

print("Исходный Series:")
print(s)

mean_val = s.mean()
print(f"Среднее значение непустых элементов: {mean_val}")

s_filled = s.fillna(mean_val)

print("Series после заполнения NaN средним значением:")
print(s_filled)
```
