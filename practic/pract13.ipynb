```python
import pandas as pd

s = pd.Series([100, 200, 300], index=['2024-03-10', '2024-03-11', '2024-03-12'])

print("Исходный Series со строковыми индексами:")
print(s)

s.index = pd.to_datetime(s.index)

print("Series после преобразования индексов в DatetimeIndex:")
print(s)

print(f"Тип данных индекса после преобразования: {type(s.index)}")
```
