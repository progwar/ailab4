```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

s = pd.Series(np.random.randint(50, 151, 30), index=pd.date_range('2024-03-01', periods=30, freq='D'))

plt.plot(s.index, s.values, marker='o')
plt.title('График значений')
plt.xlabel('Дата')
plt.ylabel('Значение')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
