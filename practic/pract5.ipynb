```python
import pandas as pd

animals = ['cat', 'dog', 'rabbit', 'parrot', 'fish']
s = pd.Series(animals)

print("Исходный Series:")
print(s)

print(f"Тип данных до преобразования: {s.dtype}")

s_category = s.astype('category')

print("Series после преобразования в category:")
print(s_category)
```
