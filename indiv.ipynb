```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = """Дата,Продажи
2024-07-01,120
2024-07-02,150
2024-07-03,170
2024-07-04,160
2024-07-05,180
"""

with open('sales.csv', 'w', encoding='utf-8') as f:
    f.write(data)

df = pd.read_csv('sales.csv', index_col='Дата', parse_dates=['Дата'])
s = df['Продажи']

print("Исходные данные:")
print(s)

s_rolling = s.rolling(2).mean()

print("Скользящее среднее:")
print(s_rolling)

plt.figure(figsize=(12, 6))

plt.plot(s.index, s.values, marker='o', label='Реальные продажи')

plt.plot(s_rolling.index, s_rolling.values, marker='s', color='red', linestyle='--', label='Скользящее среднее')

plt.title('Продажи в интернет-магазине', fontsize=16)

plt.xlabel('Дата', fontsize=12)
plt.ylabel('Продажи', fontsize=12)

plt.grid(True, alpha=0.3)

plt.legend(loc='upper left', fontsize=12)

plt.tight_layout()

plt.show()
```
